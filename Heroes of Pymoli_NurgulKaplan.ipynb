```python
# **Top Spenders**

# * Identify the the top 5 spenders in the game by total purchase value, then list (in a table):
#   * SN
#   * Purchase Count
#   * Average Purchase Price
#   * Total Purchase Value
```

```python
Top_spenders = Purchase_df.groupby(['SN'])
Top_spenders.head(10)


Top_spenders_df = pd.DataFrame({
             "Purchase count": Top_spenders["Price"].count(),
              "Average Purchase Price":Top_spenders["Price"].mean(),
              "Total Purchase Value": Top_spenders['Price'].sum()})
Top_spenders_df.head(10)

Top_spenders_df = Top_spenders_df.sort_values("Total Purchase Value",ascending=False)
Top_spenders_df.head()
```

```python
Top_spenders_df['Average Purchase Price'] = Top_spenders_df['Average Purchase Price'].map("${:,.2f}".format)

Top_spenders_df['Total Purchase Value'] = Top_spenders_df['Total Purchase Value'].map("${:,.2f}".format)
Top_spenders_df.head()
```

```python
# **Most Popular Items**

# * Identify the 5 most popular items by purchase count, then list (in a table):
#   * Item ID
#   * Item Name
#   * Purchase Count
#   * Item Price
#   * Total Purchase Value
```

```python
Popular_items = Purchase_df.groupby(['Item ID','Item Name'])
Popular_items.head()


Popular_items_df = pd.DataFrame({
            "Purchase Count":Popular_items['Price'].count() ,
              "Item Price": Popular_items['Price'].sum()/Popular_items['Price'].count(),
                "Total Purchase Value": Popular_items['Price'].sum()
})



Popular_items_df.head()
```
